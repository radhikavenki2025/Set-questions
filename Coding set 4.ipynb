```python
#4.Count the Frequency of Each Character in a String
##Question: Write a Python program to count the frequency of each character in "programming".

word="programming"
frequency={}

for char in word:
    if char in frequency:
        frequency[char] +=1
    else :
        frequency[char] = 1

print(frequency)
        
       
    
```
